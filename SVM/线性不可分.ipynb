```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVC
```

```python
data = sio.loadmat('ex6data2.mat')
print(data.keys())
```

```python
X,y = data['X'],data['y']
print(X.shape,y.shape)

def plot_data():
    plt.scatter(X[:,0],X[:,1],c = y.flatten(),cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('y1')

plot_data()
```

```python
svc1 = SVC(C = 1,kernel = 'rbf', gamma = 1)
svc1.fit(X,y.flatten())
```

```python
svc1.score(X,y.flatten())
```

```python
def plot_boundary(model):  #绘制决策边界
    x_min,x_max = 0,1
    y_min,y_max = 0.4,1
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,500),np.linspace(y_min,y_max,500))
    z = model.predict(np.c_[xx.flatten(),yy.flatten()])
    zz = z.reshape(xx.shape)
    plt.contour(xx,yy,zz)

plot_boundary(svc1)
plot_data()
```

```python
svc50 = SVC(C = 1,kernel = 'rbf', gamma = 50)
svc50.fit(X,y.flatten())
```

```python
svc50.score(X,y.flatten())
```

```python
plot_boundary(svc50)
plot_data()
```
